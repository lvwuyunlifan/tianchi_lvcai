```python
import os 
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
label = pd.read_csv('label.csv')
```

```python
for i in range(61):
    li = label[label['label'] == i]
    imagenum = li['label'].count()
    print('第%d个，总共有有%d个图片'%(i, imagenum))
    imagelist = np.array(li['img_path']).tolist()
    img_path_gen, label_gen = [], []
    if imagenum < 100:
        for imagefile in imagelist:
            path, imagename = os.path.split(imagefile)
            for count in range(0,2):
                im = Image.open(imagefile)
                im = im.convert('RGB')
                im_blur = im.filter(ImageFilter.GaussianBlur)  # 模糊化
                im_sharp = im.filter(ImageFilter.UnsharpMask)  # 锐化增强
                im_detail = im.filter(ImageFilter.DETAIL)  # 细节增强
                im_smooth = im.filter(ImageFilter.SMOOTH)  # 平滑滤波
                if count == 0:
                    im_retate = im.rotate(30, Image.BICUBIC,1)  # 逆旋转30
                if count == 1:
                    im_retate = im.rotate(60, Image.BICUBIC,1)  # 逆旋转60
                    
                img_path_gen.append('./generate/' + str(count)+'ib_'+imagename)
                img_path_gen.append('./generate/' + str(count)+'is_'+imagename)
                img_path_gen.append('./generate/' + str(count)+'idetail_'+imagename)
                img_path_gen.append('./generate/' + str(count)+'ismooth_'+imagename)
                img_path_gen.append('./generate/' + str(count)+'irotate_'+imagename)
                label_gen.extend([int(i),int(i),int(i),int(i),int(i)])
                
                
                im_blur.save('./generate/' + str(count)+'ib_'+imagename)        
                im_sharp.save('./generate/' + str(count)+'is_'+imagename)  
                im_detail.save('./generate/' + str(count)+'idetail_'+imagename)            
                im_smooth.save('./generate/' + str(count)+'ismooth_'+imagename)            
                im_retate.save('./generate/' + str(count)+'irotate_'+imagename)

        label_dict = {'img_path':img_path_gen, 'label':label_gen}
        label_gen = pd.DataFrame(label_dict)
        label = label.append(label_gen)
        label['label'] = label[['label']].astype('int64')
        # print(label)
        
    if 100 <= imagenum <= 200:
        for imagefile in imagelist:
            path, imagename = os.path.split(imagefile)
            im = Image.open(imagefile)
            im = im.convert('RGB')
#             im_blur = im.filter(ImageFilter.GaussianBlur)  # 模糊化
            im_sharp = im.filter(ImageFilter.UnsharpMask)  # 锐化增强
            im_detail = im.filter(ImageFilter.DETAIL)  # 细节增强
#             im_smooth = im.filter(ImageFilter.SMOOTH)  # 平滑滤波
            im_retate = im.rotate(30, Image.BICUBIC,1)
            
#             img_path_gen.append('./generate/' + 'ib_'+imagename)
            img_path_gen.append('./generate/' + 'is_'+imagename)
            img_path_gen.append('./generate/' + 'idetail_'+imagename)
#             img_path_gen.append('./generate/' + 'ismooth_'+imagename)
            img_path_gen.append('./generate/' + 'irotate_'+imagename)
            label_gen.extend([int(i),int(i),int(i)])


#             im_blur.save('./generate/' + 'ib_'+imagename)        
            im_sharp.save('./generate/' + 'is_'+imagename)  
            im_detail.save('./generate/' + 'idetail_'+imagename)            
#             im_smooth.save('./generate/' + 'ismooth_'+imagename)            
            im_retate.save('./generate/' + 'irotate_'+imagename)

        label_dict = {'img_path':img_path_gen, 'label':label_gen}
        label_gen = pd.DataFrame(label_dict)
        label = label.append(label_gen)
        label['label'] = label[['label']].astype('int64')
        
    if 200 <= imagenum <= 300:
        for imagefile in imagelist:
            path, imagename = os.path.split(imagefile)
            im = Image.open(imagefile)
            im = im.convert('RGB')
#             im_blur = im.filter(ImageFilter.GaussianBlur)  # 模糊化
            im_sharp = im.filter(ImageFilter.UnsharpMask)  # 锐化增强
            im_detail = im.filter(ImageFilter.DETAIL)  # 细节增强
#             im_smooth = im.filter(ImageFilter.SMOOTH)  # 平滑滤波
            
#             img_path_gen.append('./generate/' + 'ib_'+imagename)
            img_path_gen.append('./generate/' + 'is_'+imagename)
            img_path_gen.append('./generate/' + 'idetail_'+imagename)
#             img_path_gen.append('./generate/' + 'ismooth_'+imagename)
#             img_path_gen.append('./generate/' + 'irotate_'+imagename)
            
            label_gen.extend([int(i),int(i)])


#             im_blur.save('./generate/' + 'ib_'+imagename)        
            im_sharp.save('./generate/' + 'is_'+imagename)  
            im_detail.save('./generate/' + 'idetail_'+imagename)            
#             im_smooth.save('./generate/' + 'ismooth_'+imagename)            
            # im_retate.save('./generate/' + 'irotate_'+imagename)

        label_dict = {'img_path':img_path_gen, 'label':label_gen}
        label_gen = pd.DataFrame(label_dict)
        label = label.append(label_gen)
        label['label'] = label[['label']].astype('int64')
    
    if 300 <= imagenum <= 400:
        for imagefile in imagelist:
            path, imagename = os.path.split(imagefile)
            im = Image.open(imagefile)
            im = im.convert('RGB')
            # im_blur = im.filter(ImageFilter.GaussianBlur)  # 模糊化
            im_sharp = im.filter(ImageFilter.UnsharpMask)  # 锐化增强
#             im_detail = im.filter(ImageFilter.DETAIL)  # 细节增强
#             im_smooth = im.filter(ImageFilter.SMOOTH)  # 平滑滤波
            
            # img_path_gen.append('./generate/' + 'ib_'+imagename)
            img_path_gen.append('./generate/' + 'is_'+imagename)
#             img_path_gen.append('./generate/' + 'idetail_'+imagename)
#             img_path_gen.append('./generate/' + 'ismooth_'+imagename)
            # img_path_gen.append('./generate/' + 'irotate_'+imagename)
            label_gen.extend([int(i)])


            # im_blur.save('./generate/' + 'ib_'+imagename)        
            im_sharp.save('./generate/' + 'is_'+imagename)  
#             im_detail.save('./generate/' + 'idetail_'+imagename)            
#             im_smooth.save('./generate/' + 'ismooth_'+imagename)            
            # im_retate.save('./generate/' + 'irotate_'+imagename)

        label_dict = {'img_path':img_path_gen, 'label':label_gen}
        label_gen = pd.DataFrame(label_dict)
        label = label.append(label_gen)
        label['label'] = label[['label']].astype('int64')
        
    if imagenum > 1500:
        li2=li.sample(frac=0.5)
        rowlist=[]
        li2_index = li2.index
        print(li2_index)
        for indexs in range(len(li2_index)):
            if indexs <=(imagenum - 1500):
                rowlist.append(li2_index[indexs])
        label=label.drop(rowlist,axis=0)
label.to_csv('./AgriculturalDisease_trainingset/label_gen.csv')
```
